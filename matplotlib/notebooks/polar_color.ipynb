```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
N = 10000
r = 2*np.random.rand(N)
theta = 2*np.pi*np.random.rand(N)
colors = theta
ax = plt.subplot(111, polar=True)
c = ax.scatter(theta, r, c=colors, s=1, cmap='gist_rainbow')
ax.set_axis_off()
```

```python
N = 70
r = 2
cmap = 'gist_rainbow'
alpha = 0.5

x = np.random.rand(N)
y = np.random.rand(N)

area =  2000 * np.random.rand(N) ** r
colors = np.random.rand(N)

ax = plt.subplot(111)
c = ax.scatter(x, y, c=colors, s=area, cmap=cmap)
c.set_alpha(alpha)
ax.set_axis_off()
```
